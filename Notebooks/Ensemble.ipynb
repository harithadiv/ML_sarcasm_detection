```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score, roc_curve, roc_auc_score, classification_report, confusion_matrix
pd.set_option('display.max_colwidth', -1)
```

```python
bert = "bert_out.csv"
lstm = "lstm_out.csv"
svm = "svm_out.csv"
lr = "lr_out.csv"
gold = "sarcasm_test.csv"
```

```python
bert_out = pd.read_csv(bert,engine='python', header=None)[0].values
lstm_out = pd.read_csv(lstm,engine='python', header=None)[0].values
svm_out = pd.read_csv(svm,engine='python').iloc[:,1].values
lr_out = pd.read_csv(lr,engine='python', header=None)[0].values
gold_out = pd.read_csv(gold,engine='python')['label'].values
comments = pd.read_csv(gold,engine='python')['comment'].values
```

```python
assert len(lstm_out) == 251594 
assert len(bert_out) == 251594 
assert len(lr_out) == 251594
assert len(svm_out) == 251594  
assert len(gold_out) == 251594
```

```python
ensemble_out = []

for x in range(251594):
  votes = {0:0, 1:0}
  votes[bert_out[x]] += 2
  votes[lstm_out[x]] += 1
  votes[lr_out[x]] += 1
  votes[svm_out[x]] += 1
  ensemble_out.append(0 if votes[0] > votes[1] else 1)

ensemble_out = np.array(ensemble_out)
```

```python
df = pd.DataFrame({'svm':svm_out, 'lr':lr_out, 'lstm': lstm_out, 'bert': bert_out, 'ensemble': ensemble_out, 'gold': gold_out, 'comment': comments})
```

```python
print(classification_report(gold_out, ensemble_out, labels=[1,0], target_names=['Positive', 'Negative'], digits=4))
```
